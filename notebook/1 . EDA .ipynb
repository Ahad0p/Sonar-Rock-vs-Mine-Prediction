```python
# For feature index 0 (or a column name if headers exist)
sns.histplot(df[0], kde=True, bins=30, color='skyblue')
plt.title("Distribution of Feature 0")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x=df[0])
plt.title("Boxplot of Feature 0")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Choose a subset for visibility
sns.pairplot(df[[0, 1, 2, 3]])
plt.show()
```

```python
# Assuming 59 is the target column
sns.violinplot(x=df[59], y=df[0])
plt.title("Feature 0 vs Target Distribution")
plt.show()
```
